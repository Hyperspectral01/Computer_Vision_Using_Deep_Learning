```python

import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
image_path=input("Enter the image path:")
actual_image=cv2.imread(image_path)
cv2_imshow(actual_image)
```

Dimensions of the image

```python
print(actual_image.shape)
print(actual_image)
```

Resizing the image

```python
smaller_image=cv2.resize(actual_image,(512,512))
cv2_imshow(smaller_image)
```

Dimensions of the smaller image

```python
print(smaller_image.shape)
print(smaller_image.shape[0])

print(smaller_image)
```

Converting to grayscale

```python
grayscale=cv2.cvtColor(smaller_image,cv2.COLOR_RGB2GRAY)
cv2_imshow(grayscale)
```

Checking the dimensions and the color channels of the grayscale image

```python
print(grayscale.shape)
print(grayscale)
```

Basically the image is a 2-D array of one channeled color specifying the brightness of the white colour by values between 0 and 255

**Trying different techniques**

**cv2.erode()**

```python
kernel=np.ones((7,7),np.uint8)
eroded=cv2.erode(smaller_image,kernel)
cv2_imshow(eroded)
print(eroded.shape)

print("ACTUAL DATA")
print(eroded)
```

**cv2.dilate()**

```python
kernel=np.ones((7,7),np.uint8)
dilated=cv2.dilate(smaller_image,kernel)
cv2_imshow(dilated)
print(dilated.shape)

print("ACTUAL DATA")
print(smaller_image)
print("MODIFIED DATA")
print(dilated)
```

**cv2.medianblur()**

```python
# kernel=np.ones((7,7),np.uint8)
median_blurred=cv2.medianBlur(grayscale,9)
cv2_imshow(median_blurred)
print(median_blurred.shape)

print("ACTUAL DATA")
print(grayscale)
print("MODIFIED DATA")
print(median_blurred)
```

**cv2.blur()**

```python
blurred=cv2.blur(grayscale,(5,5))
cv2_imshow(blurred)
print(blurred.shape)

print("ACTUAL DATA")
print(grayscale)
print("MODIFIED DATA")
print(blurred)
```

**cv2.GausianBlur()**

```python
gausian_blurred=cv2.GaussianBlur(grayscale,(5,5),0)
cv2_imshow(gausian_blurred)
print(gausian_blurred.shape)

print("ACTUAL DATA")
print(grayscale)
print("MODIFIED DATA")
print(gausian_blurred)
```
